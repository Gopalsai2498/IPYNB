Then the derivative is zero, this creates a problem in the learning process.

In Gradient Descent, we need to calculate derivative and backpropagate the errors. If the derivatives are close to 0, the learning becomes extremely slow. This is called **Vanishing Gradient's** problem.

Sigmoid and tanh have this Vanishing Gradient problem that's why they make the learning process very slow. Then they came up with the new function called ReLU




Then they came up with the new function called ReLU which is extremely simple function. For example, if the value is less than 0 then the output is 0 if the output is more than 0 then the output is the same as that value. If we feed 2, we get 2 as an output. If we feed -1 as an input, we get 0 as an output.

The guideline is for hidden layers, ReLU is more popularly used function because it is computationally very effective. For sigmoid and tanh, we have to do some kind of computation. ReLU is very, very lightweight function 

For hidden layers if you are not sure of which activation function to be used, just use ReLU as default choice.

ReLU also has Vanishing Gradient problem, because if the value is less than 0, the derivative is again 0 - 0/1 leads to 0 derivative. Then there is another function called **Leaky ReLU**




The function **Leaky ReLU**, has some sort of linear line when the values are less than 0. It is still trying to reduce the value close to 0 but the equation is instead of 0 is 0.1x.




Which activation function to use?

The answer is unclear. If the output is binary, simply use sigmoid. In the hidden layer, mostly use ReLU or Leaky ReLU. These work on a trial and error basis.

```python
#Python function for sigmoid
import math

#euhlers - math.exp()

def sigmoid(x):
    #converts any given value b/w 0 and 1
    return 1/(1+math.exp(-x))
```

```python
#tanh

def tanh(x):
    #converts any given value between -1 and +1
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```
