How to do profit and loss calculations for revenue and expenses tables:



The data is 2-dimensional, at obs level products and at the column levels, it has years 

If we look it at matrices level, the data looks like this:



Matrix is a grid numbers arranged in rows and columns 

Let's take another case:**Matrix Multiplication**:



Dot product is multiplying two vectors when data is in single row and single column 

```python
##Python exercise:
import numpy as np
#numpy allows to do matrix manipulation
```

```python
#create multidimensional array using numpy
revenue = np.array([[180,200,220], [24,36,40], [12,18,20]])
expenses = np.array([[80,90,100], [10,16,20],[8,10,10]])
```

```python
revenue
```

```python
expenses
```

```python
#numpy allows to perform math operations between multidimensional array
profit = revenue - expenses
profit
```

```python
units = np.array([[30,40,50], [5,10,15], [2,5,7]])
price = np.array([1000,400,1200])
```

```python
units
```

```python
price
```

**Broadcasting**
If the shape of two matrices are not the same, it will try to expand it 

```python
#this concept is called broadcasting in numpy 
price * units
```

**Dot product**

```python
np.dot(price,units)
```

```python
30*1000+5*400+2*1200
```
