**Errors**


Now we sum up all the errors for all the samples by applying log loss function.

**Epoch**



We noted the log loss of 4.31 after the first epoch, which is going through all the training samples once.

Then our goal is to backpropagate this log loss error to minimize the error by adjusting the w1 and w2.  In this process we might have to run multiple epochs until we get the correct value of w1 and w2 which produces lesser error than 4.31. Remember when we train neural network, we might have to run multiple epochs.

Now we need to update the w1 and w2 weights in such a way that the log loss can be reduced which is less than 4.31 here.

**Derivatives**

Well, how to adjust w1 and w2?

We need to subtract some value from the present w1; need to adjust it basically either subtract or add some value. So what is that value?




Derivative of loss compared to w1 indicates that how much the log loss changes for a given change in w1.

Learning rate is limiting the derivative function.

Both the plains are like U-shaped function. We start at a random point where the loss is very high which is represented with star in the plot and we are trying to gradually move to a point where loss is minimal and that point is nothing but a Global Minima, the red circle in the plot.
