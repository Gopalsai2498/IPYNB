Then what is Partial Derivative?

**Partial Derivatives**

For example, when we have 2 variables x & y, to find the partial derivative.



In order to calculate Partial Derivative for function(f) with respect to x variable, we make 'y' variable (y^2) zero and find the derivative of x^3 which is 3x^2 and that's the partial derivative for x variable.

The same process is applied to y variable - so the partial derivative of function (f) with respect to 'y' variable, we make x^3 zero and find a derivative of y^2 which is 2y and that is the partial derivative for y variable
