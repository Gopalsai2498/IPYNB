Based on age and affordability, we are trying to predict whether a person will buy insurance or not.



Specifically if we look at an obs with 47 years of age and affordability being 1, a person will most likely buy insurance as per data

We need to come up with a prediction function which is y=f(x)

**Logistic Regression**



In Logistic Regression, based on the age and the affordability, calaculated weighted sum and applied sigmoid function.

Here we need to figure out the optimal w1 and w2. How?

We go through the training sample one-by-one. Initially, we try to start with random weights for w1 and w2. For example, we we will initialize w1 and w2 with 1 values. But it can be zero, it can be any random value.

**sample 1**



We initiallized the w1 and w2 to be 1 and we fed the first sample into neural network. And we got the predicted output as 0.99 and the actual value for this sample is 0. So it is obviously making an error. So we will find out the error. One of the ways to find the error is mean absolute error.  


**sample 2**

Take the second sample and pass it to the neural network.



Repeat this process till the end of samples in the data set


**sample end**



The process feeding from first sample to the end sample to the neural network is called **forward pass**.

**Loss & Cost Function**



And then we will accumulate all those errors. Take average of all the errors, which is Mean Absolute error. MAE is called **Cost function**. Individiual errors like error 1, error 2, etc. is called **Loss**. 


Sometimes Loss & Cost function are used as synonyms. Individual errors are called Loss. The cumulative error or MAE is called Cost Function. 


**Epoch**



After first epoch, the MAE is 10.02. Once the neural network went through all the training samples, it means it finished the single round of forward passing of all the samples in the given training set, it is called **one epoch**. So *one epoch* means going through all the training samples once.

If epoch is set to be 5, it means, the neural network will go through all the training samples 5 times.

in binary crossentropy the formula caculates log(yhat) -> log(0) or log(1); log(y-yhat) -> log(1-1) -> log(0), will all create problems in binary crossentropy calculations:

a) whenever predicted value is 0, then log(0) will result in -inf values, and 
b) whenever predicted value is 1, then log(1) will result in 0 values

In order to overcome the above issues, need to come up with an epsilon

**log loss**



```python
#log loss function

def log_loss(y,y_pred):
    epsilon = 1e-15
    y_pred_new = [min(i,1-epsilon) for i in y_pred]
    y_pred_new = [max(i,epsilon) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    logloss = -np.mean(y*np.log(y_pred_new)+(1-y)*np.log(1-y_pred_new))
    return logloss
```

```python
log_loss(y_true,y_pred)
```
